```python
from nilearn import plotting
from nilearn import datasets
from nilearn import image
import numpy as np
from scipy.sparse import random
from nilearn.datasets import load_mni152_template
from nilearn.datasets import fetch_surf_fsaverage
from nilearn.datasets import load_mni152_brain_mask
from enigmatoolbox.plotting import plot_subcortical
from neuromaps.datasets import fetch_fslr
import matplotlib

#%pip install surfplot
from surfplot import Plot

#%pip install brainspace
from brainspace.datasets import load_parcellation

```

```python
## Voxel-based statistical map threshold and overlaid over a T1-weighted template data, 
## with a single axial slice shown. Made using the the ortho2 function from the neurobase package.


## T1 version
rsn = datasets.fetch_atlas_smith_2009()['rsn10']
img = image.index_img(rsn, 0)
plotting.plot_stat_map(img, 
                       threshold=2, 
                       display_mode="z", 
                       cut_coords = True,
                       colorbar=True, 
                       cmap= "inferno", 
                       resampling_interpolation="continuous", 
                       output_file='data/python/fig2a.png')

```

```python
## A voxel-level cortical parcellation overlaid on T1-weighted MRI data shown in 9-slice axial orientation. 

dataset_ho = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')

plt = plotting.plot_roi(dataset_ho.maps, 
                        display_mode='z',
                        cmap = 'inferno',
                        cut_coords = 4,
                        output_file='data/python/fig2b.png')

```

```python
## Using the ENIGMA toolbox
## A 3D rendering of 7 bilateral subcortical regions (aseg atlas) where a statistical value has been assigned to each region. 

# Generate 14 random values to plot on subcortical surfaces (excluding ventricles)
np.random.seed(seed=100)
subcort_data = random(16,1, density=1).toarray()

plot_subcortical(array_name=subcort_data, 
                 size=(800, 400), 
                 embed_nb=True,
                 cmap='viridis', 
                 color_bar=True, 
                 color_range=(0, 1), 
                 interactive=False, zoom=1.25,
                 screenshot=True, 
                 filename='data/python/fig2e.png',
                 transparent_bg = False)

#crop to only show one hemisphere, as the enigma toolbox does not allow this natively
from PIL import Image 
im = Image.open('data/python/fig2e.png')
im = im.crop( (355, 0, 800, 400) ) 
im.save('data/python/fig2e.png') # saves the image


```

```python
# Medial and lateral views of a ROI atlas displayed on inflated template cortical surface where a statistical
## value has been assigned to each region.

surfaces = fetch_fslr()
lh, rh = surfaces['inflated']
p = Plot(lh, rh, layout='grid')


# add schaefer parcellation (no color bar needed)
lh_parc, rh_parc = load_parcellation('schaefer')

#create variable with random numbers for each cortical parcel, can instead use your own data too
lh_unique = np.unique(lh_parc)
lh_unique = lh_unique[1:201] #remove medial wall
rh_unique = np.unique(rh_parc)
rh_unique = rh_unique[1:201] #remove medial wall

for i in range(lh_unique.shape[0]):
    rd = np.random.random()
    rh_parc = np.where(rh_parc==rh_unique[i], rd, rh_parc)
    rd = np.random.random()
    lh_parc = np.where(lh_parc==lh_unique[i], rd, lh_parc)



p.add_layer({'left': lh_parc, 'right': rh_parc}, color_range = [0,1],
            as_outline=False, cbar=True)

#p.add_layer({'left': lh_parc, 'right': rh_parc}, color_range = [0,1], cmap = 'Greys',
#            as_outline=True, cbar=False) #optional - add border line to regions

fig = p.build()

fig.savefig('data/python/fig2c.png', dpi=600)
```

```python
## lateral views of vertex-level data displsayed on a individuals white matter surface.
fsaverage = datasets.fetch_surf_fsaverage()

destrieux_atlas = datasets.fetch_atlas_surf_destrieux()
parcellation = destrieux_atlas['map_left']


#lateral
plotting.plot_surf_roi(fsaverage.white_left, 
                   parcellation,
                   cmap='viridis',
                   bg_map=fsaverage['sulc_left'],
                   colorbar=False, view = 'lateral',
                   bg_on_data = True, darkness=0.1, 
                   engine = 'matplotlib', 
                   output_file="temp1.png")
#medial 
plotting.plot_surf_roi(fsaverage.white_left, 
                   parcellation,
                   cmap='viridis',
                   bg_map=fsaverage['sulc_left'],
                   colorbar=False, view = 'medial',
                   bg_on_data = True, darkness=0.1, 
                   engine = 'matplotlib',
                   output_file="temp2.png")

#Combine lateral and medial output .png into one file. Alternatively can do as matplotlib subplots. 
from PIL import Image 
med = Image.open('temp1.png')
lat = Image.open('temp2.png')
med_lat =[np.asarray(med), np.asarray(lat)]
med_lat = np.hstack(med_lat)
med_lat = Image.fromarray(med_lat)
med_lat.save('data/python/fig2d.png' )    

import os
files = ['temp1.png', 'temp2.png']
for f in files:
    os.remove(f)
```

```python
## A weighted and undirected graph plotted on top, left and front views of a template of a brain rendering. 

from nilearn.connectome import ConnectivityMeasure

import numpy as np
from nilearn import surface


atlas = destrieux_atlas
coordinates = []
labels = destrieux_atlas['labels']
for hemi in ['left', 'right']:
    vert = destrieux_atlas['map_%s' % hemi]
    rr, _ = surface.load_surf_mesh(fsaverage['pial_%s' % hemi])
    for k, label in enumerate(labels):
        if "Unknown" not in str(label):  # Omit the Unknown label.
            # Compute mean location of vertices in label of index k
            coordinates.append(np.mean(rr[vert == k], axis=0))

coordinates = np.array(coordinates)  # 3D coordinates of parcels

# We now make a synthetic connectivity matrix that connects labels
# between left and right hemispheres.
n_parcels = len(coordinates)

from scipy.sparse import random


#generate random correlation matrix 
corr = random(150,150, density=0.05).toarray()
corr = (corr*corr.T)/2
corr = corr/np.max(corr)

plt = plotting.plot_connectome(corr, 
                               coordinates, node_size=10, 
                               node_color='lightslategrey', 
                               edge_threshold="90%", 
                               edge_cmap='viridis', 
                               edge_vmin=0, 
                               edge_vmax=1.0,
                               colorbar=True)


plt.savefig('data/python/fig2f.png', dpi=600) 
```
